```python
# fetch all data where c1=11 or c11=2
cursor.execute('select c1,c11 from sidfirst.glassdata where c1=11 or c11=2')
```

```python
# setting data in ascending order with respect to column2
cursor.execute('select * from sidfirst.glassdata order by c2')
```

```python
# for descending order just add desc at the end
```

```python
# fetch data from table for c1>160
# and also after that filtered data fetch data where c11=3
# half operation just for understanding
cursor.execute('select * from sidfirst.glassdata where c1>160')
```

```python
cursor.fetchall()
```

```python
# complete operation with mention of alias as B its important ..here we do nested query
cursor.execute("select * from (select * from sidfirst.glassdata where c1>160) as B where c11=3")
```

```python
cursor.fetchall()
```
